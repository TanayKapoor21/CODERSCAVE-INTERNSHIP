EXPLORATORY DATA ANALYSIS

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='MONTH', y='BILL_AMT', data=df1)
plt.title('Sales Trend Overtime')
plt.xlabel('Month')
plt.ylabel('Bill amount')
plt.show()
```

```python
col = 'BRD'
i = df[col].value_counts().index
v = df[col].value_counts().values
default_color = '#3498db'
fig, ax = plt.subplots()
ax.pie(v, labels=i, startangle=90, colors=[default_color])
ax.axis('equal')
plt.title(f"Pie Plot of {col}", fontsize=20)
plt.show()
```

```python
m_sales = df1.groupby('MONTH').sum()
plt.figure(figsize=(15,10))
sns.barplot(x=m_sales.index,y='BILL_AMT',data=m_sales)
plt.title("Total Sales by Month")
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(axis='y',color='brown');
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(x='STORECODE', y='BILL_AMT', hue='MBRD', data=df1)
plt.title('Store Behaviour')
plt.xlabel('STORECODE')
plt.ylabel('BILL_AMT')
plt.show()
```
